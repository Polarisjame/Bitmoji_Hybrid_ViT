```python
from PIL import Image
import os
from matplotlib import pyplot as plt
```

```python
square = Image.open(fileroot)
```

```python
fig, axes = plt.subplots(1, 4)
axes_list = []
for i in range(4):
    axes_list.append(axes[i])
for ind,file in enumerate(os.listdir(fileroot)):
    file_path = fileroot+file
    pic = Image.open(file_path)
    axes_list[ind].imshow(pic)
    if ind == 3:
        break
```

```python
import torchvision.transforms as T
normalize = T.Normalize(mean=[0.485, 0.456, 0.406],
                                    std=[0.229, 0.224, 0.225])
transforms = T.Compose([
                    T.Resize(256),
                    T.CenterCrop(224),
                    T.RandomHorizontalFlip(),
                    T.RandomRotation(90),
                    T.ToTensor(),
                    normalize
                    ])
```

```python
fig, axes = plt.subplots(1, 4)
axes_list = []
for i in range(4):
    axes_list.append(axes[i])
for ind,file in enumerate(os.listdir(fileroot)):
    file_path = fileroot+file
    pic = Image.open(file_path)
    pic = transforms(pic).permute(2, 1, 0)
    axes_list[ind].imshow(pic)
    if ind == 3:
        break
```
